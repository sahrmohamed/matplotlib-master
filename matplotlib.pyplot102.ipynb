<h1 style="color: #148f8f69; font-family: Sans Serif;"> matplotlib.pyplot Crash Course</h1>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

<h3 style="color: #148f8f69; font-family: Sans Serif;">Function Plot</h3>

```python
x1 = np.linspace(0, 5, 10)
y1 = x1**2
plt.style.use('ggplot')
plt.plot(x1, y1)
plt.title('Days Squared Chart')
plt.xlabel('Days')
plt.ylabel('Days Square')
plt.show()
```

<h3 style="color: #148f8f69; font-family: Sans Serif; font-style: italic;">Print Multiple Plots</h3>

```python
plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'r')
plt.subplot(1, 2, 2)
plt.plot(x1, y1, 'b')
plt.show()
```

<h3 style="color: #148f8f69; font-family: Sans Serif; font-style: italic;">Using Figure Objects</h3>

```python
fig_1 = plt.figure(figsize=(5, 4), dpi=100)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x1, y1, label='x/x')
axes_1.plot(x1, y1, label='x/x')
# upper right: 1, upper left: 2, lower left: 3, lower right: 4,
# or supply a tuple of x & y from lower left
axes_1.legend(loc=0)


axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x1, y1, 'r')

axes_2.text(0, 40, 'Message')
```

<h3 style="color: #148f8f69; font-family: Sans Serif; font-style: italic;">SubPlots</h3>

```python
fig_2, axes_2 = plt.subplots(figsize=(8, 4), nrows=1, ncols=3)
plt.tight_layout()
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[1].plot(x1, y1, 'red')
axes_2[0].plot(x1, y1, 'blue')
```

<h3 style="color: #148f8f69; font-family: Sans Serif; font-style: italic;">Appearance Options</h3>

```python
# Defaul colors (b: blue, g: green, r: red, c: cyan, m: magenta \
# y: yellow, b: black, w: white)
# color='0.75' creates a 75% gray
# You can use hexcodes color="#eeefff"
# You can use color names found next like this color="burlywood"


fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])
axes_3.plot(x1, y1, color='navy', alpha = .75, lw=2, ls="-.", marker='o', markersize=7,
            markerfacecolor='y', markeredgecolor='y', markeredgewidth=4)

axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

axes_3.grid(True, color='0.6', dashes=(5,2,1,2))
axes_3.set_facecolor('#FAEBD7')
```

<h3 style="color: #148f8f69; font-family: Sans Serif; font-style: italic;">Pie Chart</h3>

```python
import random

fig_6 = plt.figure(figsize=(8,5), dpi=100)
axes_6 = fig_6.add_axes([0.1,0.1,0.9,0.9])

#create a pie chart of the number or pokemon by type
types = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
         'Ground', 'Rock', 'Fighting', 'Dark','Steel', 'Electric', 'Dragon', 'Fairy',
         'Ghost', 'Ice']
poke_num = [133,109,101,98,85,77,68,66,65,60,57,54,53,51,50,50,46,40]

colors = []

for i in range(18):
    rgb = (random.uniform(0,.5),random.uniform(0,.5),random.uniform(0,.5))
    colors.append(rgb)

explode = [0] * 18
explode[0] = 0.2

wedges, texts, autotexts = plt.pie(poke_num, explode=explode, labels=types,
        colors=colors, autopct='%1.0f%%', shadow=True, startangle=140,textprops=dict(color='w'))
plt.legend(wedges, types, loc='right', bbox_to_anchor=(1,0,0.5,1))
plt.show()
```

<h3 style="color: #148f8f69; font-family: Sans Serif; font-style: italic;">3D Surfaces</h3>

```python
from mpl_toolkits import mplot3d

fig_9 = plt.figure(figsize=(8,5), dpi=100)
axes_9 = fig_9.add_axes([0.1,0.1,0.9,0.9], projection='3d')

# A scatter3D
z_3 = 15 * np.random.random(100)
x_3 = np.sin(z_3) * np.random.random(100)
y_3 = np.cos(z_3) * np.random.random(100)
axes_9.scatter3D(x_3, y_3, z_3, c=z_3, cmap='Blues')
```

```python
# A Contour
fig_9 = plt.figure(figsize=(8,5), dpi=100)
axes_9 = fig_9.add_axes([0.1,0.1,0.9,0.9], projection='3d')
def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
x_4 = np.linspace(-6, 6, 30)
y_4 = np.linspace(-6, 6, 30)

x_4, y_4 = np.meshgrid(x_4, y_4)
z_4 = get_z(x_4, y_4)

axes_9.view_init(45, 55)
axes_9.contour3D(x_4, y_4, z_4, 80, cmap='Purples')

axes_9.set_xlabel('x')
axes_9.set_ylabel('y')
axes_9.set_zlabel('z')
```

```python
# WireFrame
fig_9 = plt.figure(figsize=(8,5), dpi=100)
axes_9 = fig_9.add_axes([0.1,0.1,0.9,0.9], projection='3d')
def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
x_4 = np.linspace(-6, 6, 30)
y_4 = np.linspace(-6, 6, 30)

x_4, y_4 = np.meshgrid(x_4, y_4)
z_4 = get_z(x_4, y_4)

axes_9.view_init(45, 55)
axes_9.plot_wireframe(x_4, y_4, z_4, color='purple')

axes_9.set_xlabel('x')
axes_9.set_ylabel('y')
axes_9.set_zlabel('z')
```

```python
# A plot_surface
fig_9 = plt.figure(figsize=(8,5), dpi=100)
axes_9 = fig_9.add_axes([0.1,0.1,0.9,0.9], projection='3d')
def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
x_4 = np.linspace(-6, 6, 30)
y_4 = np.linspace(-6, 6, 30)

x_4, y_4 = np.meshgrid(x_4, y_4)
z_4 = get_z(x_4, y_4)

axes_9.view_init(45, 55)
axes_9.plot_surface(x_4, y_4, z_4, rstride=1, cstride=1, cmap='Purples', edgecolor='none')

axes_9.set_xlabel('x')
axes_9.set_ylabel('y')
axes_9.set_zlabel('z')
```
